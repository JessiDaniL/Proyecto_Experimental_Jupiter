```python
from astropy.coordinates import SkyCoord
import astropy.units as u
import numpy as np
from astropy.time import Time
from datetime import datetime, timezone, timedelta
```

```python
#Datos experimentales

#Como los datos están tomados en Bogotá D.C, hay que adicionar 5 horas para la correción del tiempo
bogota = timezone(timedelta(hours=-5))

#1ra toma 09/02/2026 (6:56 pm)

AR1 = '7h 9m 56s'
DEC1 = '+22d 51m 21s'

#2da toma 28/02/2026 (7:14 pm)
AR2 = '7h 6m 8s'
DEC2 = '+22d 59m 36s'

#3ra toma 02/03/2026 (6:26 pm)
AR3 = '7h 4m 32s'
DEC3 = '+23d 2m 45s'

#4ta toma 11/03/2026 (7:02 pm)

AR4 = '7h 4m 2s'
DEC4 = '+22d 58m 28s'

#5ta toma 19/03/2026 (8:02 pm)
AR5 = '7h 4m 39s'
DEC5 = '+23d 1m 55s'

#6ta toma 26/03/2026 (7:06 pm)
AR6 = '7h 5m 50s'
DEC6 = '+23d 3m 6s'

#7ta toma 6/04/2026 (7:12 pm)
AR7 = '7h 8m 59s'
DEC7 = '+22d 55m 51s'

#8va toma 30/04/2026 (6:47 pm)
AR8 = '7h 20m 31s'
DEC8 = '+22d 31m 48s'
```

```python
coords = [
    SkyCoord(AR1, DEC1),
    SkyCoord(AR2, DEC2),
    SkyCoord(AR3, DEC3),
    SkyCoord(AR4, DEC4),
    SkyCoord(AR5, DEC5),
    SkyCoord(AR6, DEC6),
    SkyCoord(AR7, DEC7),
    SkyCoord(AR8, DEC8)
]

alpha = np.array([c.ra.rad for c in coords])
delta = np.array([c.dec.rad for c in coords])
```

```python
tiempos = Time([
    datetime(2026, 2, 9, 18, 56, 0, tzinfo=bogota),
    datetime(2026, 2, 28, 19, 14, 0, tzinfo=bogota),
    datetime(2026, 3, 2, 18, 26, 0, tzinfo=bogota),
    datetime(2026, 3, 11, 19, 2, 0, tzinfo=bogota),
    datetime(2026, 3, 19, 20, 2, 0, tzinfo=bogota),
    datetime(2026, 3, 26, 19, 6, 0, tzinfo=bogota),
    datetime(2026, 4, 6, 19, 12, 0, tzinfo=bogota),
    datetime(2026, 4, 30, 18, 47, 0, tzinfo=bogota)
])

tiempos_dias_jd = tiempos.jd
tiempos_dias_jd
```

```python
np.savez('conversion_datos_Exp.npz', alpha=alpha, delta=delta, tiempos_dias_jd=tiempos_dias_jd)
```
